# question2: IMDB's top rated 100 movies


```python
#importing libraries.
from bs4 import BeautifulSoup   
import requests
import pandas as pd
#requesting webpage server to get source code
page = requests.get("https://www.imdb.com/chart/top")   
page
#getting page content
soup = BeautifulSoup(page.content) 
soup
#getting movies name
movies_name = soup.find_all('td', class_="titleColumn")
movies_name
movies = []
for movie in movies_name:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies






```

```python
#getting rating
movie_rating = soup.find_all('td',class_="ratingColumn imdbRating")
movie_rating
movierating = []
for i in movie_rating:
    i = i.get_text().replace('\n', "")
    i = i.strip(" ")
    movierating.append(i)
movierating
```

```python
#creating dataframe
data = pd.DataFrame()
data['movies'] = movies
data['rating'] = movierating
data.head(100)
```

# Question3 : IMDB TOP RATED INDIAN MOVIES

```python
#import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page
soup = BeautifulSoup(page.content)
soup
```

```python
#movie names
movies_names = soup.find_all('td', class_="titleColumn")
movies_names
movies = []
for i in movies_names:
    i = i.get_text().replace('\n', "")
    i = i.strip(" ")
    movies.append(i)
movies

      
```

```python
#IMDB rating 
movies_rating = soup.find_all('td', class_='ratingColumn imdbRating')
movies_rating
ratings = []
for rating in movies_rating:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
data = pd.DataFrame()
data['movies'] = movies
data['rating'] = ratings
data.head(100)
```

# question 4:list of respected former presidents of india


```python
#importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
soup = BeautifulSoup(page.content)
#President Names
president_names = []
for p in soup.find_all('div',class_="presidentListing"):
    president_names.append(p.find("h3").text.split("(")[0])
president_names
```

```python
#Term of Office

term_of_office = []
for t in soup.find_all('div',class_="presidentListing"): 
    term_of_office.append(t.find("p").text.split("(")[0])
term_of_office
```

```python
data = pd.DataFrame()
data['presidents'] = president_names
data['term of office'] = term_of_office
data
```

# 9) Write a python program to scrape mentioned details from dineout.co.in 

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
soup = BeautifulSoup(page.content)

# a) Restaurant name
titles = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

```python
# b) Cuisine

cuisine = []
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text.split('|')[1])
    
cuisine
```

```python
# c) Location

Location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
Location
```

```python
 #d) Ratings

rating = []
for i in soup.find_all('div',class_="rating-txt hide"):
    rating.append(i.text)
    
rating
```

```python
# e) Image URL

image_url = []
for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
    
image_url
```
